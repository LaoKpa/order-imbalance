```python
cols = ['ts', 'entries', 'ltp', 'buy_c', 'sell_c', 'buy_vol', 'sell_vol', 'max_r', 'min_r']
```

```python
df = df[cols]
```

```python
df.head()
```

```python
df.drop(["max_r", "min_r"], axis = 1, inplace = True)
df.head(40)
```

```python
df["ts"] = pd.to_datetime(df["ts"])
```

```python
df = df.set_index("ts")
```

```python
entries_df = df["entries"].resample("1S").sum()
ltp_df = df["ltp"].resample("1S")#.max()
buy_c_df = df["buy_c"].resample("1S").sum()
sell_c_df = df["sell_c"].resample("1S").sum()
buy_vol_df = df["buy_vol"].resample("1S").sum()
sell_vol_df = df["sell_vol"].resample("1S").sum()
```

```python
df2 = pd.DataFrame()
```

```python
df2["entries"] = entries_df
```

```python
df2[["open", "high", "low", "close"]] = ltp_df
```

```python
ltp_df
```

```python
df2["buy_c"], df2["sell_c"], df2["buy_vol"], df2["sell_vol"] = buy_c_df, sell_c_df, buy_vol_df, sell_vol_df
```

```python
df2["ltp_change"] = df2["high"] - df2["low"]
```

```python
df2.drop(["open", "high", "low", "close"], axis = 1, inplace = True)
df2.head()
```

```python
df2["voi"] = df2["buy_vol"] - df2["sell_vol"]
df2["oir"] = df2["buy_vol"]/df2["sell_vol"]
```

```python
df2.to_csv("voi_oir.csv")
```
